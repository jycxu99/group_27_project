```r
library(reshape2)
install.packages("WDI")
library(WDI)
```

## 1. Get data

- First we download an updated list of inidicators.

```r
new_wdi_cache <- WDIcache()
```

### WDI Indicator search

- Next we find out the IDs for the indicators we are intersted in: GDP, life expectancy, and infant mortality.

```r
WDIsearch("gdp.*capita.*US\\$", cache = new_wdi_cache)
WDIsearch("life expectancy at birth.*total", cache = new_wdi_cache)
WDIsearch("^mortality.*rate.*infant", cache = new_wdi_cache)
```

- Finally we download the data.

```r
wdi_dat <- WDI(indicator = c("NY.GDP.PCAP.CD", "SP.DYN.LE00.IN", "SP.DYN.IMRT.IN"), start = 2000, end = 2015, extra = TRUE) 

names(wdi_dat)
```

## 2. Clean up

```r
# remove aggregate regions and NA's
wdi_dat <- subset(wdi_dat, region != "Aggregates")

# rename indicators
names(wdi_dat)[which(names(wdi_dat) == "NY.GDP.PCAP.CD")] <- "GDP_per_capita"
names(wdi_dat)[which(names(wdi_dat) == "SP.DYN.LE00.IN")] <- "life_expectancy"
names(wdi_dat)[which(names(wdi_dat) == "SP.DYN.IMRT.IN")] <- "infant_mortality"

wdi_dat
```
